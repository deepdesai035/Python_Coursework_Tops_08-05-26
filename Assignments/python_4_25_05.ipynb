13. From the string: "HP-Laptop " Perform:

· Remove extra spaces

e Convert to lowercase

o Replace "-" with space

```python
name = "HP-Laptop "

# remove extra space

name = name.replace(" ", "")
print(name)

# lowercase 

name = name.lower()
print(name)

# replcae - with spcae 

name = name.replace("-" , " ")
print(name)

```

15. Create two sets: A = (1,2,3,4) B = (3,4,5,6) Find:

· Union

e Intersection

· Difference A - B

```python
a = {1,2,3,4,5} 
b = {3,4,5,6,7}

# union 

c = a.union(b)
print(c)

# intersection 

d = a.intersection(b)
print(d)

#difference A- b 

e = a.difference(b)
print(e)
```

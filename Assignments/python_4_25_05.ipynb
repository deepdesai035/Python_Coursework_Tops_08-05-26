15. Create two sets: A = (1,2,3,4) B = (3,4,5,6) Find:

· Union

e Intersection

· Difference A - B

```python
a = {1,2,3,4,5} 
b = {3,4,5,6,7}

# union 

c = a.union(b)
print(c)

# intersection 

d = a.intersection(b)
print(d)

#difference A- b 

e = a.difference(b)
print(e)
```

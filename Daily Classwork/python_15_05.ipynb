# String and String slicing ,dicing 


```python
city = "Mumbai city"

print(city[0])

print(city[1])
print(city[-2])

# Mumbai city
```

```python
# SLicing 

text = "python is programming language "
print(text[0:6])
print(text[6:])
print(text[:])
print(text[:12])
```

```python
# capitalzation , lower case , Capitalization

text = "python is a programming language"
print(text.upper())
print(text.lower())
print(text.capitalize())
```

```python
# Replace 

product_name = "iphone 14pro max"

print(product_name.replace("iphone", "Samsung"))
```

```python
#split 

product = "name of the person is deep desai "
print(product.split( " "))
```

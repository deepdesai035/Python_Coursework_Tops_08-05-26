```python
# Q2
# Take input for marks in 3 subjects as string,
# convert them into float,
# and print average marks.
```

```python
marks1 = float(input("Enter the marks of the Subject 1 :- "))
marks2 = float(input("Enter the marks of the Subject 2 :- "))
marks3 = float(input("Enter the marks of the Subject 3 :- "))

average_marks = (marks1 + marks2 + marks3) / 3
print(f"The average marks of the 3 subjects is :- {average_marks}")
```

```python
# Q3
# Create a variable:
# product = "   HP-Laptop   "
# Perform:
# - remove spaces
# - convert to lowercase
# - replace "-" with space
# Print final output.
```

```python
product = "   HP-Laptop   "

product1 = product.replace(" ", "")
print(product1)
product2 = product1.lower()
print(product2)
product3 = product2.replace("-"," ")
print(f"THe final output is {product3}.")
```

```python
# Q5
# Create a string:
# course = "ArtificialIntelligence"
# Print:
# a) First 10 characters
# b) Last 5 characters
# c) Every second character using slicing
```

```python
course = "ArtificialIntelligence"
print(course[:10])
length = len(course)
print(course[(length-5):])
print(course[::2]) # step 2 
```

```python
def discount(disc , originalp):
    if disc > 100:
        return 
    else:    
        final_price = originalp - originalp*disc/100
        return final_price


originalprice = int(input("the the original price :- "))

discountp = int(input("Enter the percentage of discount you want to give :--"))

Total = discount(discountp , originalprice)

if Total == None:
    print("invalid discount percetage")

else:

    print(f"final Price :- {Total}")
```

```python
#6 

# count how many numbers are greater than 6 in a list 

list1 = [213,235,65,6,3,23,67778,6,83,1,5,3,2]
count = 0 
for i in list1:
    if i > 6:
        count = count+1

print(count)
```

```python
#9 

# largest in a list 
list1 = [113,234,5,7,4,23,5,56,7,578,33,123,646,453,243]

large = list1[0]

for i in list1:
    if i > large:
        large = i 

print(large)
```

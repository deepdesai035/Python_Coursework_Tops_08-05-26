```python
def countdown(n):

    if n==n:
        print("done")
        return

    print(n)

    countdown(n-1)

countdown(5)        
```

```python
def factorial(n):
    fact
    if n==1:
        print(fact)
        return
    fact = 1 
    fact = fact *n

    factorial(n-1) 

factorial(5)       
```

```python
def factorial(n):
    if n==1:
        return 1

    return n * factorial(n-1)

factorial(5)        
```

```python
def sum(n):
    if n==0:
        return 0

    return n + sum(n-1)

sum(1999)        
```

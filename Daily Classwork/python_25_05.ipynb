### Dictionary 

```python
product = {"laptop": 50000, "Mobile": 20000, "tablet": 15000}

print(product)
```

```python
# Access 

print(product["laptop"])
print(product["Mobile"])
print(product["tablet"])


# ADD 

product["Tablet"] = 15000
print(product)

# update 
product["Laptop"] = 45000
print(product)

#delete
product["Mobile"] = 15000
print(product)

del product["Mobile"]
print(product)

#keys 

print(product.keys())

print(product.values())


# items 

print(product.items())



```

```python
# tuples method

numbers = (10,20,30,40,50)

#count ,--- > total number of occurences of a specific value in the tuple

print(numbers.count(20))

# index --- > it returns the index of the first occurence of a specific value in the tuple

print(numbers.index(30))
```

```python
location = {28,50,72,28,50}
location2 = {28,50,72,28,50,89,23,21,54,30}

# adding elemtent

location.add(100)
print(location)

# removing element ( if not fpund it will raise an error)
# location.remove(38)

# discard element ---> no errror if no elemnt found 

location.discard(38)
print(location)

# pop element (usally removes the rendom element from the set)

location.pop()
print(location)

# union 

union = location.union(location2)
print(union)

#intersection

intersection = location.intersection(location2)
print(intersection)

# difference 

difference = location.difference(location2)
print(difference)
```
